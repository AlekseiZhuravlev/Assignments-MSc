## Task 2.1

a.) Prove the following identities
$$
\begin{align}
R_Y(-\theta)R_Y(+\theta)=R_Y(+\theta)R_Y(-\theta)=I
\end{align}
$$

Proof: Using $\cos(-x)=\cos(x)$ and $\sin(-x)=-sin(x)$
$$
\begin{align}
R_Y(-\theta)=
\begin{pmatrix} 
    \cos \left(-\frac\theta2\right) & -\sin \left(-\frac\theta2\right) \\ 
    \sin \left(-\frac\theta2\right) & \cos \left(-\frac\theta2\right)
\end{pmatrix} 
=\begin{pmatrix} 
    \cos \left(\frac\theta2\right) & \sin \left(\frac\theta2\right) \\ 
    -\sin \left(\frac\theta2\right) & \cos \left(\frac\theta2\right)
\end{pmatrix}
=R_Y(\theta)^{\top}
\end{align}
$$
From Task 1.1 we know that $R_Y(\theta)$ is unitary and thus the identities holds.

b.) Prove the following identity
$$
\begin{align}
R_Y^2\left(\frac\theta2\right):=R_Y\left(\frac\theta2\right)R_Y\left(\frac\theta2\right)=R_Y\left(\theta\right)
\end{align}
$$

Proof: Using $2\cos(x)\sin(x)=\sin(2x)$ and $\cos^2(x)-\sin^2(x)=cos(2x)$
$$
\begin{align}
R_Y\left(\frac\theta2\right)R_Y\left(\frac\theta2\right)&=
\begin{pmatrix} 
    \cos \left(\frac\theta2\right) & -\sin \left(\frac\theta2\right) \\ 
    \sin \left(\frac\theta2\right) & \cos \left(\frac\theta2\right)
\end{pmatrix}
\begin{pmatrix} 
    \cos \left(\frac\theta2\right) & -\sin \left(\frac\theta2\right) \\ 
    \sin \left(\frac\theta2\right) & \cos \left(\frac\theta2\right)
\end{pmatrix} \\
&=\begin{pmatrix} 
    \cos^2 \frac\theta2 - \sin^2 \frac\theta2 & -2\cos \frac\theta2\sin \frac\theta2 \\ 
    2\cos \frac\theta2\sin \frac\theta2 & \cos^2 \frac\theta2 - \sin^2 \frac\theta2
\end{pmatrix} \\
&=\begin{pmatrix} 
    \cos \theta & -\sin \theta \\ 
    \sin \theta & \cos \theta
\end{pmatrix}
=R_Y\left(\theta\right)
\end{align}
$$

c.) Prove the following identity
$$
\begin{align}
XR_Y\left(-\theta\right)X=R_Y\left(\theta\right)
\end{align}
$$

Proof: 
$$
\begin{align}
XR_Y\left(-\theta\right)X&=
\begin{pmatrix} 
    0 & 1 \\ 
    1 & 0
\end{pmatrix}
\begin{pmatrix} 
    \cos \left(\frac\theta2\right) & \sin \left(\frac\theta2\right) \\ 
    -\sin \left(\frac\theta2\right) & \cos \left(\frac\theta2\right)
\end{pmatrix}
\begin{pmatrix} 
    0 & 1 \\ 
    1 & 0
\end{pmatrix} \\&=
\begin{pmatrix} 
    -\sin \left(\frac\theta2\right) & \cos \left(\frac\theta2\right) \\
    \cos \left(\frac\theta2\right) & \sin \left(\frac\theta2\right) 
\end{pmatrix}
\begin{pmatrix} 
    0 & 1 \\ 
    1 & 0
\end{pmatrix} \\
&=\begin{pmatrix} 
    \cos \left(\frac\theta2\right) & -\sin \left(\frac\theta2\right) \\ 
    \sin \left(\frac\theta2\right) & \cos \left(\frac\theta2\right)
\end{pmatrix}
=R_Y\left(\theta\right)
\end{align}
$$

## Task 2.3

Import standard libraries

```python
import numpy as np
from qiskit import Aer, transpile
from qiskit import QuantumCircuit
from qiskit import QuantumRegister
from qiskit import ClassicalRegister
from qiskit.circuit.library.standard_gates import *
from qiskit.visualization import plot_histogram
```

Compute rotation angles

```python
thtX1 = 2 * np.arcsin(np.sqrt(0.25))
thtY1X1 = 2 * np.arcsin(np.sqrt(0.70))
thtY1X0 = 2 * np.arcsin(np.sqrt(0.60))
thtZ1Y1X1 = 2 * np.arcsin(np.sqrt(0.90))
thtZ1Y1X0 = 2 * np.arcsin(np.sqrt(0.70))
thtZ1Y0X1 = 2 * np.arcsin(np.sqrt(0.55))
thtZ1Y0X0 = 2 * np.arcsin(np.sqrt(0.10))
```

Define quantum registers

```python
qregX = QuantumRegister(1, name='x')
qregY = QuantumRegister(1, name='y')
qregZ = QuantumRegister(1, name='z')
```

Define operator $A$

```python
opA = QuantumCircuit(qregX, qregY, qregZ, name='A')
# 1 Ry gate on x
opA.ry(thtX1, qregX)
# 2 controlled Ry gates on y
opA.cry(thtY1X1, control_qubit=qregX, target_qubit=qregY)
opA.x(qregX)
opA.cry(thtY1X0, control_qubit=qregX, target_qubit=qregY)
opA.x(qregX)
# 4 multiple controlled Ry gates on z
opA.mcry(thtZ1Y1X1, [qregX[0], qregY[0]], qregZ[0])
opA.x(qregX)
opA.mcry(thtZ1Y1X0, [qregX[0], qregY[0]], qregZ[0])
opA.x(qregX)
opA.x(qregY)
opA.mcry(thtZ1Y0X1, [qregX[0], qregY[0]], qregZ[0])
opA.x(qregY)
opA.x(qregX)
opA.x(qregY)
opA.mcry(thtZ1Y0X0, [qregX[0], qregY[0]], qregZ[0])
opA.x(qregY)
opA.x(qregX)
```

Add measurment

```python
qcirc = QuantumCircuit(qregX, qregY, qregZ, name='Bayes net')
qcirc.append(opA, [qregX, qregY, qregZ])
qcirc.measure_all()
```

Helper function for running circuits on simulator

```python
simulator = Aer.get_backend('aer_simulator')
def runCircuit(circ, shots=10_000):
    transp = transpile(circ, simulator)
    IBMjob = simulator.run(transp, shots=shots)
    result = IBMjob.result()
    counts = result.get_counts(transp)
    return counts
```

a.) Run the following snippet

```python
nshots = 10_000
counts = runCircuit(qcirc, nshots)
for key in sorted(counts.keys()):
    print (key, counts[key] / nshots)
plot_histogram(counts, bar_labels=False)
```

Since quantum measurements are probabilistic, multpile runs return different results

c.) Implement $A^{\dagger}$

* $A\in \mathbb R^{3\times 3} \Rightarrow A^{\dagger}=A^{\top}$
* $A=U_1\cdot U_2\cdots U_m\Rightarrow A^{\top}=U^{\top}_m\cdots U^{\top}_2\cdot U^{\top}_1$
* $X^{\top}=X, ~ R_Y(\theta)^{\top}=R_Y(-\theta),~ CR_Y(\theta)^{\top}=CR_Y(-\theta),~ CCR_Y(\theta)^{\top}=CCR_Y(-\theta)$

Thus, $A^{\dagger}$ can be implemented by switching the order of single gates used for $A$ and negating the arguments of the $R_Y$-gates

```python
opAdagger = QuantumCircuit(qregX, qregY, qregZ, name='A_inv')

opAdagger.x(qregX)
opAdagger.x(qregY)
opAdagger.mcry(-thtZ1Y0X0, [qregX[0], qregY[0]], qregZ[0])
opAdagger.x(qregY)
opAdagger.x(qregX)
opAdagger.x(qregY)
opAdagger.mcry(-thtZ1Y0X1, [qregX[0], qregY[0]], qregZ[0])
opAdagger.x(qregY)
opAdagger.x(qregX)
opAdagger.mcry(-thtZ1Y1X0, [qregX[0], qregY[0]], qregZ[0])
opAdagger.x(qregX)
opAdagger.mcry(-thtZ1Y1X1, [qregX[0], qregY[0]], qregZ[0])
opAdagger.x(qregX)
opAdagger.cry(-thtY1X0, control_qubit=qregX, target_qubit=qregY)
opAdagger.x(qregX)
opAdagger.cry(-thtY1X1, control_qubit=qregX, target_qubit=qregY)
opAdagger.ry(-thtX1, qregX)
```

a.) Estimate the probabilities for the chain to end up in states $D$,$E$ or $F$

State transition probability is $\frac12$ and thus we estimate the probablities to be 
$$
\begin{align}
P(X_2=D)&=P(X_2=D, X_1=B)+P(X_2=D, X_1=C)=P(X_2=D|X_1=B)P(X_1=B)=\frac12\cdot\frac12=\frac14 \\
P(X_2=E)&=P(X_2=E|X_1=B)P(X_1=B)+P(X_2=E|X_1=C)P(X_1=C)=\frac12\cdot\frac12+\frac12\cdot\frac12=\frac12 \\
P(X_2=F)&=P(X_2=F|X_1=C)P(X_1=C)=\frac12\cdot\frac12=\frac14
\end{align}
$$
