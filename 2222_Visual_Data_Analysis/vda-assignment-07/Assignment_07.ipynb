d) Please calculate the accumulated power generation for each administrative district and plant type

```python
import pandas as pd

# load power generation dataset
df = pd.read_csv('districts/power_generation.csv')

# show the data
df
```

```python
# calculate the accumulated power generation for each administrative district and plant type

# group by administrative district and plant type and sum the power generation
df_grouped = df.groupby(['district', 'energyType'])[['netRatePower']].sum()

# show the data
df_grouped.sort_values(by=['district', 'netRatePower'], ascending=[True, False]).head(15)
```

e) Both datasets use the same IDs for the administrative districts. Merge the two datasets so that
the entries are correctly matched by administrative districts for the plant type wind.

```python
# remove rows with district = 'nan'
df = df[df['district'] != 'nan']

df_wind = df[df['energyType']=='wind'].copy()

df_wind
```

```python
df_wind.dropna(subset=['district'], inplace=True)

# remove fractional part in the district column
df_wind['district'] = df_wind['district'].astype(float).astype(int).astype(str)

# merge the two datasets
df_merged = pd.merge(df_wind, gdf, left_on='district', right_on='AGS_0')

df_merged
```

f) Plot a map that encodes the generated wind power for each administrative district. Include a
legend that tells the reader how to interpret the colors

```python
# create geopandas object
gdf_merged = gpd.GeoDataFrame(df_merged)

# plot the data
gdf_merged.plot(column='netRatePower', legend=True).set_title('Generated wind power for each administrative district')
```

g) Currently, the colormap is not very meaningful because there are two districts producing much
more power than all other districts. Recreate the plot with optimized utilization of the colorscale
range by clipping the upper bound.

```python
# clip the dataset to the 95th quantile
# set the upper bound to the 95th quantile
upper_bound = df_merged['netRatePower'].quantile(0.95)

# set values that are above the upper bound to the upper bound
df_merged_clipped = df_merged.copy()
df_merged_clipped['netRatePower'] = df_merged_clipped['netRatePower'].clip(upper=upper_bound)

gdf_merged_clipped = gpd.GeoDataFrame(df_merged_clipped)
# plot the data
gdf_merged_clipped.plot(column='netRatePower', legend=True).set_title('Generated wind power for each administrative district')
```

h) Employ geopandas to compute the size of each administrative district in Germany, and create a
plot of the generated wind power, normalized by area

```python
# compute the size of each administrative district in Germany
gdf_merged_clipped['area'] = gdf_merged_clipped['geometry'].area

# compute the normalized power generation
gdf_merged_clipped['normalized_power'] = gdf_merged_clipped['netRatePower'] / gdf_merged_clipped['area']

# multiply by 10**6
# gdf_merged_clipped['normalized_power'] = gdf_merged_clipped['normalized_power'] * 10**6

# set the data > 95th quantile to the 95th quantile
gdf_merged_clipped['normalized_power'] = gdf_merged_clipped['normalized_power'].clip(upper=gdf_merged_clipped['normalized_power'].quantile(0.95))

# plot the data
gdf_merged_clipped.plot(column='normalized_power', legend=True).set_title('Generated wind power, normalized by area')
```
