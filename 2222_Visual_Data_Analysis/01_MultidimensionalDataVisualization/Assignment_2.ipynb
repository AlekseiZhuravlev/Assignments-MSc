```python
df_wine = pd.read_csv('winequality-red.csv', sep=";")
df_wine
```

```python
sns.histplot(df_wine, x="quality").set_title('Wine quality')
```

The range of score is from 3 to 8. The most common score is 5 and 6.

```python
df_wine['quality_bin'] = df_wine.quality.apply(lambda x: 'low' if x in [3,4] else 'medium' if x in [5,6] else 'high')
df_wine.drop('quality', axis=1, inplace=True)
```

```python
sns.histplot(df_wine['quality_bin']).set_title('Wines by quality bin')
```

```python
df_wine_low = df_wine[df_wine['quality_bin'] != 'medium'].copy()
```

```python
sns.pairplot(df_wine_low, hue='quality_bin')
```

```python
from IPython.display import Image
Image(filename='pairplot.png')
```

I think that the attributes that best distinguish between high and low quality wines are:
- alcohol
- sulphates
- volatile acidity
- citric acid
- pH
Because they separate the data the most on the plots.

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import StandardScaler
import numpy as np

X = df_wine_low.drop('quality_bin', axis=1)
y = df_wine_low['quality_bin']

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

selector = SelectKBest(f_classif, k=5)
selector.fit(X_scaled, y)

scores = -np.log10(selector.pvalues_)
scores /= scores.max()

attributes_scores = np.transpose(np.vstack((X.columns, scores)))

for i in sorted(attributes_scores, key=lambda x: x[1], reverse=True)[:5]:
    print(i)
```

```python
from scipy.spatial.distance import pdist, squareform
from scipy.stats import spearmanr
import matplotlib.pyplot as plt

def distance_consistency(x, y):
    x = x.values.reshape(-1, 1)
    y = y.values.reshape(-1, 1)
    # print(np.hstack((x, y)))
    d = pdist(np.hstack((x, y)), metric='euclidean')
    return np.mean(d) / np.std(d)

def spearman_correlation(x, y):
    return spearmanr(x, y)[0]

def plot_consistency(df, metric):

    consistency = np.zeros((len(df.columns), len(df.columns)))
    for m, c1 in enumerate(df.columns):
        for n, c2 in enumerate(df.columns):
            consistency[m, n] = metric(df[c1], df[c2])

    ax = sns.heatmap(consistency, xticklabels=df.columns, yticklabels=df.columns, cmap='RdBu_r')
    ax.set_title('Distance consistency')
    ax.set_xlabel('Attribute')
    ax.set_ylabel('Attribute')
    plt.show()

df_wine_low_filtered_pre = df_wine_low_filtered.drop('quality_bin', axis=1)
plot_consistency(df_wine_low_filtered_pre, distance_consistency)
```
