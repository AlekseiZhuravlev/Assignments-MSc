```python
import igraph
import matplotlib.pyplot as plt
import sklearn
import numpy as np
```

# TASK 1



```python
class MessagePassingFramework:
    def __init__(self, graph, update_function = None, aggregate_function = None):
        self.graph = graph
        self.update_function = update_function
        self.aggregate_function = aggregate_function
        
    def set_update_function(self, update_function):
        self.update_function = update_function
        
    def set_aggregate_function(self, aggregate_function):
        self.aggregate_function = aggregate_function
        
    def message_passing_graph(self, propty):
        # creating a copy of the graph to avoid 
        # modifying properties during message passing
        
        graph_copy = self.graph.copy()
        
        for i, vertex_copy in enumerate(graph_copy.vs):
            vertex_graph = self.graph.vs[i]
            vertex_graph[propty] = self.update_function(
                vertex_copy[propty],
                self.aggregate_function(vertex_copy, propty)
            )
```

```python
# aggregate and update functions

def mean_aggregate(vertex, propty):
    neighborhood = vertex.neighbors()
    neighb_props = [v[propty] for v in neighborhood]
    return np.mean(neighb_props)

def mean_update(vertex_pty, agg_pty):
    return (vertex_pty + agg_pty) / 2
```

```python
# testing on a small graph

graph = igraph.Graph(n=4, edges=[[0, 1], [1, 2], [2, 3], [1, 3]])
graph.vs['val'] = [10, 19, 7, 1]

msg_pass = MessagePassingFramework(graph, update_function=mean_update, aggregate_function=mean_aggregate)
print(graph.vs['val'])

msg_pass.message_passing_graph('val')
print(graph.vs['val'])

msg_pass.message_passing_graph('val')
print(graph.vs['val'])
```
