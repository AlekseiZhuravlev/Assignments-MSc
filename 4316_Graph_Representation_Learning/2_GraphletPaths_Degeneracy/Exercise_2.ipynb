```python

def depth_first_search(G, v, parent, path_length, path_sequence):

    if len(path_sequence) == path_length:
        return [path_sequence]
    
    nodes_to_traverse = v.neighbors().copy()
    for i in path_sequence + [parent]:
        if i in nodes_to_traverse:
            nodes_to_traverse.remove(i)
            
    paths_list = []
    for i in nodes_to_traverse:
        curr_path = path_sequence.copy()
        curr_path.append(i)
        paths = depth_first_search(G, i, v, path_length, curr_path)
        if paths:
            paths_list += paths
            
    return paths_list

def print_path_sequence(path_sequence):
    for path in path_sequence:
        output = ''
        for node in path:
            output += f'{node["name"]}-'
        print(output[:-1])
    
def find_all_paths(g):
    simple_paths = []
    
    for v in g.vs:
        simple_paths += depth_first_search(g, v, None, len(g.vs), [v])
        
    return len(simple_paths) // 2, simple_paths
 
def make_graph(n, edges):
    g = igraph.Graph(n=n, edges=edges)
    vertex_names = 'abcdefghkjk'
    for i in range(n):
        g.vs[i]['name'] = vertex_names[i]
    return g
```

```python
igraph.config['plotting.backend'] = 'matplotlib'

for g in [
    make_graph(n=4, edges=[[0, 1], [1, 2], [2, 3], [1, 3]]),
    make_graph(n=4, edges=[[0, 1], [1, 2], [2, 3], [3, 0], [1, 3]]),
    make_graph(n=5, edges=[[0, 1], [0, 3], [1, 4], [1, 2], [2, 4], [2, 3]]),
    make_graph(n=5, edges=[[0, 1], [0, 3], [0, 4], [1, 3], [1, 2], [2, 3]]),
    make_graph(n=5, edges=[[0, 1], [0, 2], [0, 4], [1, 2], [1, 3], [2, 3], [2, 4], [3, 4]])
    
]:
    n_paths, simple_paths = find_all_paths(g)
    
    igraph.plot(g)
    print_path_sequence(simple_paths)
    print(f'unique simple paths: {n_paths}')
    
    plt.show()
    plt.clf()
```

```python
import pathlib
import numpy as np

data_path = pathlib.Path.cwd().parent / 'Exercise_1' / 'ogbn-arxiv.pickle'
g = igraph.Graph.Read_Pickle(data_path)
g = g.as_undirected()
```

```python
import random

def sampled_hcs(graph, vertex, sample_vertices):
    if not len(sample_vertices):
        return 1
    
    closeness_centrality = 0 
    distances = g.distances(source = [vertex], target = sample_vertices)

    for i in distances[0]:
        closeness_centrality += 1/i
    closeness_centrality /= len(sample_vertices)
    
    return closeness_centrality
    
def select_landmark_vertices(g, alpha):
    return random.sample(list(g.vs), int(len(g.vs) * alpha))
```

```python
source_vs = g.vs.select(label_eq=12)

for v in source_vs[:5]:
    print(f'vertex {v.index}, (alpha, mean hcs, std): ', end='')
    for alpha in [0.001, 0.01, 0.1, 1]:
        hcs_list = []
        
        for i in range(10):
            landmark_vs = select_landmark_vertices(g, alpha)
            if v in landmark_vs:
                landmark_vs.remove(v)

            hcs = sampled_hcs(g, v, landmark_vs)
            hcs_list.append(hcs)
            
        print(f'({alpha}, {np.mean(hcs_list):.4f}, {np.std(hcs_list):.4f})', end = ', ')
    print()
```
